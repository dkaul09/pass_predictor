```python
events = sb.events(match_id=3869151) 
print(events)
```

```python
passes = events[events['type'] == 'Pass'].copy()

# Clean up the data
# 1. Fill NaN values in 'pass_outcome' column: NaN means completed pass
passes['pass_outcome'] = passes['pass_outcome'].fillna('Completed')

# 2. Create a new column for 'pass_completed' (True/False)
passes['pass_completed'] = passes['pass_outcome'] == 'Completed'
pd.set_option('future.no_silent_downcasting', True)


# 3. Handle missing values in 'under_pressure' column: Fill with False
passes['under_pressure'] = passes['under_pressure'].fillna(False)

# 4. Remove unnecessary columns: Keeping only relevant ones for analysis
passes_clean = passes[['location', 'pass_end_location', 'under_pressure', 'pass_outcome', 'pass_completed']]
```
