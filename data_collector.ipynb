```python
import pandas as pd 
from statsbombpy import sb
from mplsoccer import Pitch
```

```python
sb.competitions()
```

```python
sb.matches(competition_id=43, season_id=106)
```

```python
selected_columns = [
    'location',
    'pass_end_location',	
    'pass_length',      
    'pass_angle',      
    'pass_outcome',    
    'under_pressure',  
    'play_pattern',     
    'pass_height',     
]


matches = sb.matches(competition_id=43, season_id=106)
all_pass_events = []

for match_id in matches['match_id']:

  events = sb.events(match_id=match_id)
  passes = events[events['type'] == 'Pass']
  passes['pass_outcome'] = passes['pass_outcome'].fillna('Completed')
  passes['under_pressure'] = passes['under_pressure'].fillna('FALSE')
  passes = passes[passes['pass_outcome'].isin(['Completed', 'Incomplete'])]
  passes = passes[selected_columns]
  all_pass_events.append(passes)

all_pass_data = pd.concat(all_pass_events, ignore_index=True)
output_file = "world_cup_pass_data.csv"
all_pass_data.to_csv(output_file, index=False)

print(f"Total matches processed: {len(matches)}")
print(f"Total pass events collected: {len(all_pass_data)}")
print(f"Pass data saved to: {output_file}")

```
